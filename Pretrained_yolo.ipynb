```python
def sort_filenames_by_number(filenames):
    def extract_number(filename):
        # Extract the numeric part from the filename
        return int(filename.split("_")[-1].split(".")[0])

    # Sort the filenames using the extract_number function as the key
    sorted_filenames = sorted(filenames, key=extract_number)

    return sorted_filenames

# Example usage
pat = '/content/Hacktech/runs/detect/train/labels'
filenames = os.listdir(pat)

sorted_files = sort_filenames_by_number(filenames)
```

```python
vidname=(os.path.basename(vidpath)).split('.')[0]
```

```python
import re
import os
tt=[]
num=1
for i in  sorted_files:
  p=os.path.join(pat,i)
  with open(p,'r') as r:
    for item in r:
      xx=[ x for x in item.split()]
      # print(item,len(xx),p)
      xmin = xx[1]
      xmax = xx[3]
      ymin = xx[2]
      ymax = xx[4]
      if len(item.split()) <=5:
        tt.append([num,item[0],xmin,xmax,ymin,ymax,int('0')])
      else:
        tt.append([num,item[0],xmin,xmax,ymin,ymax,int(item.split()[-1])])
    num+=1
```

```python
ft=open("/content/scene2_05.txt","w")
for i in tt:
  try:
    ft.write(str(vidname)+" "+str((i)[0])+" "+" ".join([str(x) for x in i[1:-1]])+" "+str((i)[-1])+"\n")
  except:
        ft.write('0000000')
```

```python
def postprop(filename):
    # Open the file for reading and writing
    with open(filename, 'r+') as file:
    # Read all lines from the file
        lines = file.readlines()

        # Iterate through each line
        for i in range(len(lines)):
            # Split the line into a list of values
            values = lines[i].split()
            # print(values[2])

            # Check if the third item is not in [5, 7, 2]
            if int(values[2]) not in [5, 7, 2]:
                # Change the third item to 0
                # print(values[2])
                values[2] = '3'
            else:
                if (values[2]=='2'):
                    values[2] = '0'
                elif (values[2]=='5'):
                    values[2] = '3'
                elif (values[2]=='7'):
                    values[2] = '1'



                # Update the line in the list
                lines[i] = ' '.join(values) + '\n'

        # Move the file cursor to the beginning
        file.seek(0)

        # Write the updated lines back to the file
        file.writelines(lines)

        # Truncate the remaining content (in case the new content is shorter than the old content)
        file.truncate()
```
