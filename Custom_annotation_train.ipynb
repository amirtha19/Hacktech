```python
def sort_filenames_by_number(filenames):
    def extract_number(filename):
        # Extract the numeric part from the filename
        return int(filename.split("_")[-1].split(".")[0])

    # Sort the filenames using the extract_number function as the key
    sorted_filenames = sorted(filenames, key=extract_number)

    return sorted_filenames

# Example usage
pat = '/content/Hacktech/runs/detect/train3/labels'
filenames = os.listdir(pat)

sorted_files = sort_filenames_by_number(filenames)
# print(sorted_files)
```

```python
import os
img_width=960
img_height=540
tt=[]
num=1
for i in sorted_files:
  p=os.path.join(pat,i)
  # print(p)
  with open(p,'r') as r:
    for item in r:
      xx=[ x for x in item.split()]
      xmin = max(0, (float(xx[1]) - 0.5 * float(xx[3])) * img_width)
      xmax = min(img_width, (float(xx[1]) + 0.5 * float(xx[3])) * img_width)
      ymin = max(0, (float(xx[2]) - 0.5 * float(xx[4])) * img_height)
      ymax = min(img_height, (float(xx[2]) + 0.5 * float(xx[4])*img_height))
      if len(item.split()) <=5:
        tt.append([num,item[0],xmin,xmax,ymin,ymax,int('0')])
      else:
        tt.append([num,item[0],xmin,xmax,ymin,ymax,int(item.split()[-1])])
    num+=1
```

```python
ft=open("/content/resorted.txt","w")
for i in tt:
  try:
    ft.write(str((i)[0])+" "+" ".join([str(x) for x in i[1:-1]])+" "+str((i)[-1])+"\n")
  except:
        ft.write('0000000')
```
